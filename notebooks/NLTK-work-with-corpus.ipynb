for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)

```python
# CREATE LIST OF SENTENCES
macbeth_sentences = gutenberg.sents("shakespeare-macbeth.txt")
# FIND THE MAXIMUM LENGTH
longest_sent = max(len(s) for s in macbeth_sentences)
# PAIR THE STRING WITH THE MAXIMUM LENGTH
print([s for s in macbeth_sentences if len(s) == longest_sent])
```

```python
from nltk.corpus import brown
brown.categories()
```

```python
religion = brown.words(categories="religion")
rel_fdist = nltk.FreqDist(w.lower() for w in religion)
rel_fdist.most_common(20)
rel_words = ["god", "jesus", "christ", "holy"]
for r in rel_words:
    print(r + ": ", rel_fdist[r], end=", ")
```

```python
news = brown.words(categories="news")
news_fdist = nltk.FreqDist(w.lower() for w in news)
for n in news:
    if n.startswith("wh"):
        print(n + ": ", news_fdist[n], end=", ")
```

```python
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ["america", "citizen"]
    if w.lower().startswith(target))
cfd.plot()
```

```python
from nltk.corpus import udhr
languages = ["Chickasaw", "English", "German_Deutsch", "Greenlandic_Inuktikut", "Hungarian_Magyar", "Ibibio_Efik"]
cfd = nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages
    for word in udhr.words(lang + "-Latin1"))
cfd.plot(cumulative=True)
```
